##**1.- Identificación del problema**
**¿De qué forma puede Spotify diferenciarse de su competencia siguiendo una estrategia basada en datos?**
En la actualidad, se estima que casi el 90% de los datos existentes en internet fueron creados en los últimos años, y de estos, gran parte permanecen sin ser analizados. se ha llegado a un punto tal que, los datos han sido catalogados como el petróleo del siglo XXI, resaltando las oportunidades que pueden obtenerse de su exploración. 
Para este proyecto, continuaremos centrándonos en la industria de streaming de música como se hizo en el módulo anterior, más específicamente, en Spotify, una de las empresas más resilientes ante la entrada de nuevos competidores y que, ante este panorama competitivo, se ha visto forzada a buscar nuevas formas de mantenerse en la punta de los ingresos generados en el sector.

##**2.- Investigación**
Como puede observarse en el gráfico anterior, la industria de los servicios de streaming de música está fuertemente concentrada en unas pocas empresas, de hecho, cinco de ellas representan el 80% del mercado y las más grandes: Spotify y Apple Music, acaparan poco más de la mitad de estos ingresos. 
Dada esta alta concentración, es interesante analizar cómo estas empresas compiten entre sí y los punteros, se esfuerzan por diferenciarse al ofrecer un servicio que consta de lo mismo en su raiz (escuchar música en cualquier momento a través de un dispositivo). 
Como bien sintentiza Serenity Gibbons en un artículo de la revista Forbes, existen básicamente tres estrategias a seguir para tomar la delantera en essta industria:


*   **Ganar lealtad durante la fase de prueba:** Aquí se debe dejar suficiente espacio al usuario como para que interactúe con la plataforma de forma placentera sin restringirle demasiado las funcionalidades de la misma, pero al mismo tiempo, dejarle con el deseo de pasar a una suscripción con el fin de incorporar a su experiencia la totalidad de las herramientas del servicio. Aquí es fundamental recordar constantemente de todas las ventajas que el usuario puede obtener de su suscripción, sin dejar espacio a que sea únicamente el conocimiento previo el encargado de tomar la decisión final. 
*   **Precios competitivos:** Cuando el servicio es tan similar entre plataformas, un precio demasiado elevado puede simplemente alejar a los posibles usuarios, es por ello, que una estrategia bien segmentada de precios ofreciendo distintos paquetes a los clientes es la clave para que aquellos que dudan sobre la adquisición den ese paso adeltante.
*   **Diversificar la oferta:** Si un fan no encuentra a su banda favorita dentro de la plataforma, será difícil que se incorpore a un plan de pago, pero la diversificación va más allá. Facilitar la entrada a otro tipo de formatos como los podcasts o el mismo video son escenciales para una óptima diferenciación.

##**3.- Búsqueda de soluciones anteriores**
Un reciente estudio realizado en la universidad de Duke se dio a la tarea de analizar las características ofrecidas por los servicios de streaming y su importancia para los consumidores. Los resultados se resumen en el gráfico siguiente: 
Importancia características streaming.png

Como puede observarse, el precio es el factor fundamental al momento de decidir la suscripción a un servicio de streaming, como bien hemos mencionado, este servicio es casi identico en lo fundamental entre marcas, por lo que un precio demasiado elevado simplemente alejará a nuestros posibles usuarios. 
Pero la parte a destacar viene en los dos siguientes factores que más influyen en el pago de la suscripción: la creaciñon de playlists y las recomendaciones realizadas por el algoritmo. Es justamente en este último que podemos implementar una estrategia basada en datos, pues las recomendaciones serán más acertadas mientras mejores sean los modelos de recomendación que implementemos, alimentándolos con data de primera mano generada por los mismos usuarios, otorgándonos el diferenciador que estamos buscando. 
Enfocarse en otras estrategias con el objetivo de mejorar la descarga de música o las opciones para compartir la misma en distintas redes sociales o apps de mensajería sería menos fructífero acorde a este estudio, a pesar de que ello puede también complementarse perfectamente con los datos generados por los usuarios en las plataformas.
